```python
marks = [65, 70, 72, 68, 75, 80]
```

```python

df = pd.DataFrame(marks, columns=["Marks"])
print(df)
```

```python

mean_marks = df["Marks"].mean()
max_marks = df["Marks"].max()
min_marks = df["Marks"].min()

# Output
print("Mean:", mean_marks)
print("Maximum:", max_marks)
print("Minimum:", min_marks)
```

```python
##Observation: 
A noticeable difference exists between the highest and lowest marks, showing variation in performance.
```

```python
#Interpretation 
the data suggests consistent academic performance with moderate variation among students
```

```python
# Plot line graph
plt.plot(marks, marker='o')
plt.title("Line Graph of Marks")
plt.xlabel("Observation Number")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
